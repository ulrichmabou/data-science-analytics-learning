```python
import matplotlib.pyplot as plt

%matplotlib inline
```

## Simple Plots

```python
x = [1,2,3,4,5,6,7] # week days
y = [964,287,42,390,64,800,523] # views

plt.plot(x, y, 
         label='YouTube Views', 
         color='r', 
         marker='o', markerfacecolor='g', 
         linestyle='--', linewidth=3)

plt.xlabel('Week days')
plt.ylabel('Views')
plt.title('Daily YouTube Views')
plt.legend()
plt.show()
```

## Plotting multiple datasets in a single plot

```python
y_views=[534,690,258,402,724,689,352]
f_views=[123,342,700,305,406,648,325]
t_views=[202,209,176,415,824,389,550]
days=[1,2,3,4,5,6,7]

plt.plot(days,y_views,label='Youtube views',marker='o')
plt.plot(days,f_views,label='Facebook views',marker='o')
plt.plot(days,t_views,label='Twitter views',marker='o')

plt.xlabel('Days')
plt.ylabel('Views')
plt.title('Social media views for a channel')
plt.legend()

plt.xlim(1,5)
plt.ylim(100,900)

plt.grid(True, color='r', linestyle='--')

plt.savefig('img_2.png', dpi=500, facecolor='cyan')
plt.show()
```

## Scatter Plots

```python
y_views=[534,690,258,402,724,689,352]
f_views=[123,342,700,305,406,648,325]
t_views=[202,209,176,415,824,389,550]
days=[1,2,3,4,5,6,7]

plt.scatter(days,y_views,label='Youtube views',marker='o')
plt.scatter(days,f_views,label='Facebook views',marker='o')
plt.scatter(days,t_views,label='Twitter views',marker='o')

plt.xlabel('Days')
plt.ylabel('Views')
plt.title('Social media views for a channel')
plt.legend()

plt.xlim(1,5)
plt.ylim(100,900)

plt.grid(True, color='r', linestyle='--')

plt.savefig('img_3.png', dpi=500)
plt.show()
```

## Bar Charts

```python
y_views=[200,50,250,300,100,150,350]

days=[1,2,3,4,5,6,7]

plt.bar(days,y_views,label='YouTube views')

plt.xlabel('Days')
plt.ylabel('Views')
plt.title('YouTube views for a channel')
plt.legend()
plt.savefig('bar_1.png',dpi=500)
plt.show()
```

```python
y_views=[200,50,250,300,100,150,350]
fb_views=[160,200,40,240,80,120,280]

days=[1,2,3,4,5,6,7]

plt.bar(days,y_views,label='YouTube views')
plt.bar(days,fb_views,label='Facebook views')

plt.xlabel('Days')
plt.ylabel('Views')
plt.title('Social media views for a channel')
plt.legend()
plt.savefig('bar_2.png',dpi=500)
plt.show()
```

## Histograms

```python
ages=[22,55,62,45,21,22,99,34,42,4,102,110,27,48,99,84]
bins=[0,20,40,60,80,100,120]

plt.hist(ages,bins)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Current ages of people')
plt.savefig('hist.png',dpi=500)
plt.show()
```

## Pie Charts

```python
labels_1=['Facebook','Instagram','YouTube','LinkedIn']
views=[300,350,400,450]

plt.pie(views,labels=labels_1)
plt.savefig('pie1.png',dpi=500)
plt.show()
```

```python
labels_1=['Facebook','Instagram','Youtube','linked-in']
views=[300,350,400,450]

plt.pie(views,labels=labels_1,
        autopct='%1.1f%%',
        shadow=True)
plt.savefig('pie2.png',dpi=500)
plt.show()
```

```python
labels_1=['Facebook','Instagram','Youtube','linked-in']
views=[300,350,400,450]
explode_1=[0,0,0,0.2]

plt.pie(views,labels=labels_1,
        autopct='%1.1f%%',
        explode=explode_1)
plt.savefig('pie3.png',dpi=500)
plt.show()
```

```python
labels_1=['Facebook','Instagram','Youtube','linked-in']
views=[300,350,400,450]
explode_1=[0,0,0,0.2]

plt.pie(views,labels=labels_1,
        autopct='%1.1f%%',
        explode=explode_1,
        shadow=True)
plt.savefig('pie4.png',dpi=500)
plt.show()
```
