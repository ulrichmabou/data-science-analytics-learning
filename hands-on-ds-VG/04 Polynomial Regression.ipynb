```python
# Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Loading data
dataset = pd.read_csv("Gaming_data.csv")

print(dataset)

X = dataset.iloc[:, 0:1].values
print(X)

y = dataset.iloc[:, -1].values
print(y)

# dataset.describe()
```

```python
# Fitting linear regression

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

lin_reg.fit(X, y)
```

```python
# Fitting polynomial regression

from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=2) # transforming tool

X_poly = poly_reg.fit_transform(X)

print(X_poly)

```

```python
# Actual polynomial regression

lin_reg2 = LinearRegression()

lin_reg2.fit(X_poly, y)
```

```python
# Visualizing linear regression results

plt.scatter(X, y)
plt.plot(X, lin_reg.predict(X), color='red')
plt.title('Gaming Data (Linear Regression)')
plt.xlabel('Gaming Steps')
plt.ylabel('Points')
plt.show()
```

```python
# Visualizing polynomial regression results

plt.scatter(X, y)
plt.plot(X, lin_reg2.predict(poly_reg.fit_transform(X)), color='red')
plt.title('Gaming Data (Polynomial Regression)')
plt.xlabel('Gaming Steps')
plt.ylabel('Points')
plt.show()
```

```python
# Improving the model by adjusting the polynomial degree

poly_reg3 = PolynomialFeatures(degree=3)
X_poly3 = poly_reg3.fit_transform(X)
lin_reg3 = LinearRegression()
lin_reg3.fit(X_poly3, y)

plt.scatter(X, y)
plt.plot(X, lin_reg3.predict(poly_reg3.fit_transform(X)), color='red')
plt.title('Gaming Data (Polynomial Regression: degree=3)')
plt.xlabel('Gaming Steps')
plt.ylabel('Points')
plt.show()
```

```python
poly_reg4 = PolynomialFeatures(degree=4)
X_poly4 = poly_reg4.fit_transform(X)
lin_reg4 = LinearRegression()
lin_reg4.fit(X_poly4, y)

plt.scatter(X, y)
plt.plot(X, lin_reg4.predict(poly_reg4.fit_transform(X)), color='red')
plt.title('Gaming Data (Polynomial Regression: degree=4)')
plt.xlabel('Gaming Steps')
plt.ylabel('Points')
plt.show()
```

```python
# Predicting new result with linear regression

lin_reg.predict([[7.5]])
```

```python
# Predicting new result with polynomial regression - better

lin_reg4.predict(poly_reg4.fit_transform([[7.5]]))
```
