```python
# Libraries

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Sample data

labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]
```

```python
# Simple plots

plt.plot(labels, men_means, label='Men')

plt.ylabel('Scores')
plt.title('Men Scores by Group')
plt.legend()
plt.show()
```

```python
plt.plot(labels, women_means, label='Women')

plt.ylabel('Scores')
plt.title('Women Scores by Group')
plt.legend()
plt.show()
```

```python
plt.plot(labels, men_means, label='Men', color='red')
plt.plot(labels, women_means, label='Women', color='green')


plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.legend()
plt.show()
```

```python
# Bar charts

plt.bar(labels, men_means, label='Men')

plt.ylabel('Scores')
plt.title('Men scores by group')
plt.legend()
plt.show()
```

```python
plt.bar(labels, women_means, label='Women')

plt.ylabel('Scores')
plt.title('Women scores by group')
plt.legend()
plt.show()
```

```python
plt.bar(labels, men_means, label='Men')
plt.bar(labels, women_means, label='Women')

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.legend()
plt.show()
```

```python
plt.bar(labels, men_means, label='Men', width=0.35, color='green')
plt.bar(labels, women_means, label='Women', width=0.35, color='red')

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.legend(loc='upper left')
plt.show()
```

```python
# Stacked bar charts

plt.bar(labels, men_means, label='Men', width=0.35, color='green')
plt.bar(labels, women_means, label='Women', width=0.35, color='red', bottom=men_means)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.legend(loc='upper left')
plt.show()
```

```python
# Stacked bar charts with error bars

plt.bar(labels, men_means, label='Men', width=0.35, yerr=men_std)
plt.bar(labels, women_means, label='Women', width=0.35, yerr=women_std, bottom=men_means)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.legend()
plt.show()
```
