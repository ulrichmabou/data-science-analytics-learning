```python
# Libaries

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Fixing random state for reproducibility - used to generate dummy data below
np.random.seed(100)
```

```python
plt.rcdefaults() # Restore the rc params from Matplotlib's internal default style
fig, ax = plt.subplots()

# Sample data
people = ('Joel', 'Ulrich', 'Mabou', 'Fene', 'David')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))


# Simple horizontal bar chart
ax.barh(y_pos, width=performance)
plt.show()
```

```python
# Simple horizontal bar charts with labels

plt.rcdefaults()
fig, ax = plt.subplots()

# Sample data
people = ('Joel', 'Ulrich', 'Mabou', 'Fene', 'David')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))


# Simple horizontal bar chart
ax.barh(y_pos, width=performance)
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')
plt.show()
```

```python
# Horizontal bar charts with errors

plt.rcdefaults()
fig, ax = plt.subplots()

# Sample data
people = ('Joel', 'Ulrich', 'Mabou', 'Fene', 'David')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))

error = np.random.rand(len(people))


# Simple horizontal bar chart
ax.barh(y_pos, width=performance, xerr=error, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')
plt.show()
```
